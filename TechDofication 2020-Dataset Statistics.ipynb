```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Train Set Statistics

```python
Subtask1a_English=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1a/sub-task-1a-train-en.tsv",sep='\t')
Subtask1b_Bangla=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1b/sub-task-1b-train-bn.tsv",sep='\t')
Subtask1c_Gujarati=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1c/sub-task-1c-train-gu.tsv",sep='\t')
Subtask1d_Hindi=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1d/sub-task-1d-train-hi.tsv",sep='\t')
Subtask1e_Malayalam=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1e/sub-task-1e-train-ml.tsv",sep='\t')
Subtask1f_Marathi=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1f/sub-task-1f-train-mr.tsv",sep='\t')

Subtask1g_Tamil=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1g/sub-task-1g-train-ta.tsv",sep='\t')
Subtask1h_Telgu=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-1h/sub-task-1h-train-te.tsv",sep='\t')
Subtask2a_English=pd.read_csv("ICON2020-TechDofication-TraningData/sub-task-2a/sub-task-2a-train-en.tsv",sep='\t')
dff=[Subtask1a_English,Subtask1b_Bangla,Subtask1c_Gujarati,Subtask1d_Hindi,Subtask1e_Malayalam,Subtask1f_Marathi,Subtask1g_Tamil,Subtask1h_Telgu,Subtask2a_English]
```

```python
def pt(nm,ls):
    i=0
    for df in ls:
        Ytr=df.label
        labels=list(set(Ytr))
        counts=[]
        for label in labels:
            counts.append(np.count_nonzero(Ytr==label))
        plt.pie(counts,labels=labels, autopct='%1.1f%%')
        plt.title(nm[i])
        plt.show()
        i=i+1

    
```

```python
pt(['Subtask1a_English','Subtask1b_Bangla','Subtask1c_Gujarati','Subtask1d_Hindi','Subtask1e_Malayalam','Subtask1f_Marathi','Subtask1g_Tamil','Subtask1h_Telgu','Subtask2a_English'],dff)
```
